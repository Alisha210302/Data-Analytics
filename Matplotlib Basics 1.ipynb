```python
# Show all points on x and y axis
# Show markers for all data points on the plot
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate x-axis values (1 to 9)
x = np.arange(1, 10)
x
```

```python
# Generate y-axis values (1 to 18)
y = np.arange(1,19)
y
```

```python
# Select the first 9 elements from y to match x-axis length
y = y[:9] # Slicing to get elements from index 0 (inclusive) to 9 (exclusive)
y
```

```python
# Create the line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('Plot Title Here')

# Display the plot
plt.show()
```

```python
plt.plot(x, y)
plt.scatter(x,y,s=30) # Increase marker size to 50
plt.show()
```
