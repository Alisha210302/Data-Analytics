```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
x
```

```python
y
```

```python
# line plot
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line graph')
plt.xlim(0,6)
plt.ylim(0,12)
plt.show()
```

```python
# exporting a plot
plt.plot(x,y)
plt.savefig('example.png')
```
