### 5. Age Demographics

```python
# Purchase Count
bins = [0, 10, 14, 18, 22, 26, 30, 34, 38, 42, 46, 50, 54, 58, 62]
labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-64', '65-69', '70-74']

# ==============================================j
df5['Purchase ID'] = pd.cut(df5['Age'], bins=bins)
# Group by 'Age Group' and count the purchases
purchase_count = df5.groupby('Gender')['Purchase ID'].count()

print(purchase_count)
```

```python
df5
```

```python
# Define the bin edges (4-year intervals)
bins = [0, 10, 14, 18, 22, 26, 30, 34, 40]
labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
```

```python
# Use pd.cut to categorize 'Age' into bins
df5['Age Group'] = pd.cut(df5['Age'], bins=bins, labels=labels, right=False)
```

```python
# Group by 'Age Group' and calculate required metrics
age_demographics = df5.groupby('Age Group').agg(
    Purchase_Count=('Purchase ID', 'count'),
    Average_Purchase_Price=('Price', 'mean'),
    Total_Purchase_Value=('Price', 'sum'),
    Average_Purchase_Total_per_Person=('Price', lambda x: x.sum() / df5['Age Group'].value_counts()[x.index[0]])
).reset_index()
print(age_demographics)
```

```python
# Average Purchase Total per Person by Gender
avgPurchaseByGender = df5.groupby(['Gender','SN'])['Price'].sum()
avgPurchaseByGender
```

### 6. Top Spenders

#### Identify the the top 5 spenders in the game by total purchase value, then list (in atable):

```python
# SN
df5.groupby('SN')['Price'].sum().nlargest(5)
```

```python
# Purchase Count
top5Price = df5['Price'].sort_values(ascending=False).nlargest(5)
top5Price_df = top5Price.reset_index()
top5Price_df.columns = ['ID', 'Purchase Price']
top5Price_df
```

```python
# Average purchase price
average_prices=df5.groupby('SN')['Price'].mean().nlargest(5)

average_prices_df = average_prices.reset_index()
average_prices_df.columns = ['SN', 'Average Purchase Price']  # Renaming columns for clarity
average_prices_df
```

```python
# SN
totalSum = df5.groupby('SN')['Price'].sum().nlargest(5)
totalSum_df = totalSum.reset_index()
totalSum_df.columns = ['Name', 'Total price']
totalSum_df
```
