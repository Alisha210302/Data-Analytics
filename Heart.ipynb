```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
```

```python
# 1.Group on the age column
# 2. Calculate median for each numeric column on the age group 
# 3. Reset index to reset to numeric index
ages=df.groupby('age').median().reset_index()
```

```python
ages.plot.line('age','chol')
plt.show()
```

```python
# matplotlib
fig,ax = plt.subplots()
ax.plot(ages['age'],ages['chol'],ls=':',lw=1.7)
ax.set_xlabel('Age')
ax.set_ylabel('Cholestrol')
plt.show()
```

```python
# seaborn
sns.lineplot(x='age',y='chol',data=ages,linestyle=':',linewidth=1.7)
plt.xlabel('Age')
plt.ylabel('Cholestrol')
plt.show()
```

```python
thalach=df.groupby('thalach').median().reset_index()
thalach.plot.line('thalach','chol')
plt.show()
```
